## iris data ë¡œ Hierachical Clustering

```python
from sklearn.datasets import load_iris 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import math 
%matplotlib inline
```

```python
iris =load_iris()
iris_data = iris.data
iris_data_pd = pd.DataFrame(iris_data, columns = iris.feature_names)
print(iris_data_pd)
```

```python
iris_data_pd.info()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
linkage = ['complete', 'average', 'ward']
for idx, i in enumerate(linkage):
  plt.figure(idx)
  hier = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage=i)
  hier.fit(iris_data_pd.iloc[:,2:4])
  plt.scatter(iris_data_pd.iloc[:,2], iris_data_pd.iloc[:,3], c=hier.labels_)
  plt.title('Clustering :' + i)
  plt.xlabel('petal length')
  plt.ylabel('petal width')
plt.show()
```

```python
from scipy.cluster import hierarchy
```

```python
hierar = hierarchy.linkage(iris_data_pd.iloc[:, 2:4], 'complete')
plt.figure(figsize=(20, 20))
dn = hierarchy.dendrogram(hierar)
```

```python
hierar = hierarchy.linkage(iris_data_pd.iloc[:,2:4], 'single')
plt.figure(figsize=(20,20)) 
dn = hierarchy.dendrogram(hierar)
```

## fish data ë¡œ Hierachical Clustering

```python
from sklearn.datasets import load_iris 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import math 
%matplotlib inline
```

```python
fish = pd.read_csv('Fish.csv',  sep = ',')
fish
```

```python
pd.DataFrame(fish)
fish.info()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
linkage = ['complete', 'average', 'ward']
for idx, i in enumerate(linkage):
  plt.figure(idx)
  hier = AgglomerativeClustering(n_clusters=7, affinity='euclidean', linkage=i)
  hier.fit(fish.iloc[:,[1,5]])
  plt.scatter(fish.iloc[:,1], fish.iloc[:,5], c=hier.labels_)
  plt.title('Clustering :' + i)
  plt.xlabel('Weight')
  plt.ylabel('Height')
plt.show()
```

```python
hierar = hierarchy.linkage(fish.iloc[:, [1,5]], 'complete')
plt.figure(figsize=(20, 20))
dn = hierarchy.dendrogram(hierar)
```

```python
hierar = hierarchy.linkage(fish.iloc[:, [1,5]], 'single')
plt.figure(figsize=(20, 20))
dn = hierarchy.dendrogram(hierar)
```
